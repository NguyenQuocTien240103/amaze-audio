```python
import tensorflow as tf
import tensorflow_hub as hub
import numpy as np
import csv

import matplotlib.pyplot as plt
from IPython.display import Audio
from scipy.io import wavfile
```

```python
# Load the model.
model = hub.load('https://tfhub.dev/google/yamnet/1')
```

```python
# Find the name of the class with the top score when mean-aggregated across frames.
def class_names_from_csv(class_map_csv_text):
  """Returns list of class names corresponding to score vector."""
  class_names = []
  with tf.io.gfile.GFile(class_map_csv_text) as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
      class_names.append(row['display_name'])

  return class_names

class_map_path = model.class_map_path().numpy()
class_names = class_names_from_csv(class_map_path)
```

```python
print(class_names)
print("len:",len(class_names))
```

```python
def ensure_sample_rate(original_sample_rate, waveform,
                       desired_sample_rate=16000):
  """Resample waveform if required."""
  if original_sample_rate != desired_sample_rate:
    desired_length = int(round(float(len(waveform)) /
                               original_sample_rate * desired_sample_rate))
    waveform = scipy.signal.resample(waveform, desired_length)
  return desired_sample_rate, waveform
```

```python
# wav_file_name = 'speech_whistling2.wav'
wav_file_name = 'output_yell.wav'
sample_rate, wav_data = wavfile.read(wav_file_name, 'rb')
sample_rate, wav_data = ensure_sample_rate(sample_rate, wav_data)

# Show some basic information about the audio.
duration = len(wav_data)/sample_rate
print(f'Sample rate: {sample_rate} Hz')
print(f'Total duration: {duration:.2f}s')
print(f'Size of the input: {len(wav_data)}')

# Listening to the wav file.
Audio(wav_data, rate=sample_rate)
```

```python
waveform = wav_data / tf.int16.max
print("waveform",waveform)
print("len",len(waveform))
```

```python
# Run the model, check the output.
scores, embeddings, spectrogram = model(waveform)
```

```python
scores_np = scores.numpy()
spectrogram_np = spectrogram.numpy()
infered_class = class_names[scores_np.mean(axis=0).argmax()]
print(f'The main sound is: {infered_class}')
```

```python
scores, embeddings, spectrogram = model(waveform)
scores_np = scores.numpy()

frame_duration = 0.48  # mỗi frame ~0.48s

for i, frame_scores in enumerate(scores_np):
    top_class_index = frame_scores.argmax()
    top_class_name = class_names[top_class_index]
    top_score = frame_scores[top_class_index]

    start_time = i * frame_duration
    end_time = start_time + frame_duration

    print(f"Frame {i} | {start_time:.2f}s - {end_time:.2f}s")
    print(f"  Top class: {top_class_name} ({top_score:.3f})")
    print("-" * 40)
```
